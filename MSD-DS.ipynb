```python
def Grade(marks):
    if(marks>90):
        print("A+")
    elif(marks>85):
        print("A")
    elif(marks>80):
        print("B+")
    elif(marks>70):
        print("B")
    else:
        print("Fail")
```

```python
Grade(60)
```
